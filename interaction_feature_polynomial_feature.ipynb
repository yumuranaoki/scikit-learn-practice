```python
import mglearn
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np
```

```python
X, y = mglearn.datasets.make_wave(n_samples=100)
bins = np.linspace(-3, 3, 11)
which_bin = np.digitize(X, bins=bins)
print("which_bin:\n", which_bin[:5])
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
encoder.fit(which_bin)
X_binned = encoder.transform(which_bin)
print("X_binned:\n", X_binned[:5])
```

```python
X_combined = np.hstack([X, X_binned])
print(X.shape)
print(X_binned.shape)
print(X_combined.shape)
```

```python
reg = LinearRegression().fit(X_combined, y)

line = np.linspace(-3, 3, 1000, endpoint=False).reshape(-1,1)
line_binned = encoder.transform(np.digitize(line, bins=bins))
line_combined = np.hstack([line, line_binned])

plt.plot(line, reg.predict(line_combined), label='linear regression combined')
plt.legend(loc='best')
plt.plot(X[:, 0], y, 'o', c='k')
for bin in bins:
    plt.plot([bin, bin], [-3, 3], ':', c='k')
```

```python
# すべての空間で傾きが共有されているため、あまり正確ではない
print(X_combined[:5, :])
```

```python
#　ビニングしてone hot encodingを実行した後に元データをかける
X_binned_multi = X * X_binned
print(X_binned_multi[:5, :])
```

```python
X_product = np.hstack([X_binned, X_binned_multi])
print(X_product.shape)

reg = LinearRegression().fit(X_product, y)

line_product = np.hstack([line_binned,  line * line_binned])
plt.plot(line, reg.predict(line_product), label='linear regression product')
plt.legend(loc='best')
plt.plot(X[:, 0], y, 'o', c='k')
for bin in bins:
    plt.plot([bin, bin], [-3, 3], ':', c='k')
```

```python
# 多項式による特徴量抽出
from sklearn.preprocessing import PolynomialFeatures
# x^^10までの多項式を加える
poly = PolynomialFeatures(degree=10, include_bias=False)
poly.fit(X)
X_poly = poly.transform(X)

print("entries of x:\n", X[:3])
print("entries of x_poly:\n", X_poly[:3])
```

```python
reg = LinearRegression().fit(X_poly, y)

line_poly = poly.transform(line)
plt.plot(line, reg.predict(line_poly), label='polynomial linear regression')
plt.plot(X[:, 0], y, 'o', c='k')
plt.legend(loc='best')
```

```python
from sklearn.svm import SVR
```

```python
for gamma in [1, 10]:
    svr = SVR(gamma=gamma).fit(X, y)
    plt.plot(line, svr.predict(line), label='SVR gamma={}'.format(gamma))
plt.plot(X[:, 0], y, 'o', c='k')
plt.legend(loc='best')
```
