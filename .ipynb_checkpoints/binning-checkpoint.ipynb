```python
# 最適なデータ表現方法はモデルによって異なる
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
import mglearn
```

```python
import numpy as np
X, y = mglearn.datasets.make_wave(n_samples=100)
line = np.linspace(-3, 3, 1000, endpoint=False).reshape(-1,1)
print("line.shape: {}".format(line.shape))
```

```python
import matplotlib.pyplot as plt
reg = DecisionTreeRegressor(min_samples_split=3).fit(X,y)
plt.plot(line, reg.predict(line), label="decision tree")
reg = LinearRegression().fit(X, y)
plt.plot(line, reg.predict(line), label="linear regression")

plt.plot(X[:, 0], y, 'o', c='k')
plt.ylabel('regression output')
plt.xlabel('input feature')
plt.legend(loc='best')
```

```python
# 線形モデルを特徴量に対して強力にするためにビニングを行う
```

```python
#　特徴量の入力レンジを分割する
bins = np.linspace(-3, 3, 11)

# dataをそれぞれの入力レンジに分ける
which_bin = np.digitize(X, bins=bins)
print("the bin each x value is in\n", which_bin[:5])
```

```python
# ビニングしたデータをone hot encodingに変換する
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse=False)
# バリエーションを確認
encoder.fit(which_bin)
# transformによるone hot encodingの実行
X_binned = encoder.transform(which_bin)
print(X_binned[:5])
```

```python
# make_waveのdataをone hot encodingをした後に行う
line_binned = encoder.transform(np.digitize(line, bins=bins))

reg = LinearRegression().fit(X_binned, y)
plt.plot(line, reg.predict(line_binned), label='linear regression binned')

reg = DecisionTreeRegressor(min_samples_split=3).fit(X_binned, y)
plt.plot(line, reg.predict(line_binned), label='decision tree regression binned')

plt.plot(X[:, 0], y, 'o', c='k')
plt.vlines(bins, -3, 3, linewidth=1, alpha=.2)
plt.legend(loc="best")
plt.ylabel('regeression input')
plt.xlabel('input feature')
```

```python
# 線形モデルと決定木の予測が完全に一致している
# 線形モデルはビニングによって柔軟性が向上する
# 決定木はもともと分割可能であるため、むしろビニングによって柔軟性が低下する
```
